```python
import pandas as pd


publi = pd.read_csv('Advertising.csv', index_col=0) # read csv file from same diretory or publi = pd.read_csv('file - web-url') to read from a web csv file 
publi.head()
```

```python
print(publi.shape)
```

```python
x = publi[['TV','Radio','Newspaper']]
y = publi['Sales']
```

```python
x.shape
```

```python
y.shape
```

```python
import seaborn as sns


sns.pairplot(publi, x_vars=['TV','Radio','Newspaper'], y_vars=['Sales'], height=7, kind='reg')
```

```python
from sklearn.model_selection import train_test_split


x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=5)
```

## Liner regression model import

```python
from sklearn.linear_model import LinearRegression
linReg = LinearRegression()
linReg.fit(x_train,y_train)
```

```python
list(zip(['TV','Radio','Newspaper'],linReg.coef_))
```

### Predict

```python
y_pred = linReg.predict(x_test)
print(y_pred)
```

### Performance 

#### MAE(Mean Absolute Error)

```python
from sklearn import metrics
```

```python
print(metrics.mean_absolute_error(y_test,y_pred))
```

#### MSE(Mean Squared Error)

```python
print(metrics.mean_squared_error(y_test,y_pred))
```

#### RMSE(Root Mean Squared Error)

```python
import numpy as np
print(np.sqrt(metrics.mean_squared_error(y_test,y_pred))) #sqrt tira a raiz quadrada do resultado
```
